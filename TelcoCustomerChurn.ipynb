Most of the cells are either 'Yes' or 'No'. 
*   We can assign unique values (and then)
*   Replace the categorical values in all object column with Zeros (0) and ones (1).






```python
# funtion to pull up unique values from all object
def Print_unique_col_values(telecoCustomer_df):
    for column in telecoCustomer_df:
        if telecoCustomer_df[column].dtypes == 'object':
            print(f'{column} : {telecoCustomer_df[column].unique()}')
```

```python
# Recreating the dataframe without the CustomerID column
telcoCustomer_df = file.iloc[:,1:]

# Inspecting the recreated dataset
telcoCustomer_df.columns
```

```python
# calling an instance of the function - Print_unique_col_values
Print_unique_col_values(telcoCustomer_df)
```



*   We can further clean up the data by replacing all 'No phone service' and 'No internet service' with a 'No'


```python
telcoCustomer_df.replace('No internet service', 'No', inplace = True)
telcoCustomer_df.replace('No phone service', 'No', inplace=True)

Print_unique_col_values(telcoCustomer_df)
```



*   We can assign numerical values of '1' and '0' to 'Yes' and 'No' columns respectively.


```python
# List of all columns with yes, no values
yes_no_columns = ['Partner','Dependents','PhoneService','MultipleLines','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']

for col in yes_no_columns:
    telcoCustomer_df[col].replace({'Yes': 1,'No': 0},inplace=True)

# Replacing Female and Male with 1 and 0
telcoCustomer_df['gender'].replace({'Female':1,'Male':0},inplace=True)

# Printing replacements
for col in telcoCustomer_df:
    print(f'{col}: {telcoCustomer_df[col].unique()}')
```
